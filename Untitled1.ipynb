```python
import os
import sys

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class PriorMean():
    def __init__(self, m=0.0, lam=1, beta=1):
        self.m = m
        self.lam = lam
        self.beta = beta
        
    def pdf(self, x):
        return stats.norm.pdf(x, loc=self.m, scale=1/(self.beta*self.lam))
    
    def rvs(self, size=100):
        return stats.norm.rvs(loc=self.m, scale=1/(self.beta*self.lam), size=size)
        

class PriorPrecision():
    def __init__(self, a=2, b=2):
        self.a = a
        self.b = b
        
    def pdf(self, x):
        return stats.gamma.pdf(x, self.a, loc=0, scale=self.b)
    
    def rvs(self, size=100):
        return stats.gamma.rvs(self.a, loc=0, scale=self.b, size=size)


class PosteriorMean():
    def __init__(self, data=None, m=None, lam=None, beta=None):
        self.data = data
        self.m = m
        self.lam = lam
        self.beta = beta
        
        self.N = len(data)
        self.m_hat = (np.sum(data) + beta * m) / (beta + self.N)
        self.beta_hat = beta + self.N
        
    def pdf(self, x):
        return stats.norm.pdf(x, loc=self.m_hat, scale=1/(self.beta_hat*self.lam))
    
    def rvs(self, size=100):
        return stats.norm.rvs(loc=self.m_hat, scale=1/(self.beta_hat*self.lam), size=size)


class PosteriorPrecision():
    def __init__(self, data=None, a=None, b=None, beta=None, m=None):
        self.data = data
        self.a = a
        self.b = b
        self.beta = beta
        self.m = m
        
        self.N = len(data)
        self.a_hat = a + self.N / 2
        self.b_hat = b + 0.5 * (np.sum(data ** 2) + beta * (m ** 2) - (1/(beta + self.N)) * (np.sum(data ** 2) + beta * m) ** 2)
        
    def pdf(self, x):
        return stats.gamma.pdf(x, self.a_hat, loc=0, scale=self.b_hat)
    
    def mean(self):
        return stats.gamma.mean(self.a_hat, loc=0, scale=self.b_hat)
    
    def rvs(self, size=100):
        return stats.gamma.rvs(self.a_hat, loc=0, scale=self.b_hat, size=size)

    
class Likelihood():
    def __init__(self, mu=0.0, lam=1.0):
        self.mu = mu
        self.lam = lam
        
    def pdf(self, x):
        return stats.norm.pdf(x, loc=self.mu, scale=1/self.lam)
    
    def logpdf(self, x):
        return stats.norm.logpdf(x, loc=self.mu, scale=1/self.lam)
    
    def rvs(self, size=100):
        return stats.norm.rvs(loc=self.mu, scale=1/self.lam, size=size)


class PredDist():
    def __init__(self, m=None, beta=None, a=None, b=None):
        self.df = 2 * a # degree of freedom
        self.loc = m
        # self.scale = beta / (beta+1) * (a/b)
        self.scale = np.sqrt(beta / (beta+1) * (a/b))
        
    def pdf(self, x):
        return stats.t.pdf(x, self.df, self.loc, self.scale)
    
    def sf(self, x):
        """ survival function """
        return stats.t.sf(x, self.df, self.loc, self.scale)
    
    def surprise(self, x):
        mean = stats.t.mean(self.df, self.loc, self.scale)
        sf_m = self.sf(mean)
        sf_x = self.sf(x)
        return abs(sf_x - sf_m)
    
    def error(self, x):
        mean = stats.t.mean(self.df, self.loc, self.scale)
        return abs(x - mean)
    
    def rvs(self, size=100):
        return stats.t.rvs(self.df, self.loc, self.scale, size=size)
    
```

```python
class TrueDist():
    def __init__(self, mean=0.1, precision=1.4):
        self.mean = mean
        self.precision = precision
        
    def pdf(self, x):
        return stats.norm.pdf(x, loc=self.mean, scale=1/self.precision)
    
    def rvs(self, size=100):
        return stats.norm.rvs(loc=self.mean, scale=1/self.precision, size=size)
```

```python
import matplotlib.pyplot as plt
from matplotlib import animation
from matplotlib import gridspec
%matplotlib nbagg

def update(i, all_data):  
    
    plt.cla()
    ax1.cla()   
    ax2.cla()
    ax3.cla()
    ax4.cla()
    
    data = all_data[:i]
        
    ax1.set_ylim([-0.05,1])
    ax1.set_xlim([-3,3])      
    ax2.set_ylim([-0.05,1])
    # ax2.set_xlim([0,5])  
    ax3.set_xlim([0,len(data)])
    ax4.set_xlim([0,len(data)])
    
    # True distribution
    mean = 1.2
    precision = 10
    sx = np.arange(-3,3,0.01)
    sy = [TrueDist(mean, precision).pdf(x) for x in sx]
    ax1.fill_between(sx, sy, label="data distribution", color="gray", alpha=0.25)
    
    """ Data sample """
    ax1.scatter(data, np.zeros(len(data))-0.02, alpha=0.5, color="red", marker=".", label="data samples (N={})".format(len(data)))    
    ax1.vlines(x=data[-1], ymin=-1, ymax=1, alpha=0.5, color="red", label="new data")
    
    # Prior of mean
    m = 0.
    lam = 0.5
    beta = 1.
    sy = [PriorMean(m=m, lam=lam, beta=beta).pdf(x) for x in sx]
    ax1.fill_between(sx, sy, label="prior of mean", alpha=0.25)
    
    # Prior of precision
    a = 2.
    b = 2.
    sx = np.arange(0,15,0.01)
    sy = [PriorPrecision(a=a, b=b).pdf(x) for x in sx]
    ax2.fill_between(sx, sy, label="prior of precision", alpha=0.25, facecolor="blue")

    # Likelihood
    mu = 0.0
    lam = 1.0
    sx = np.arange(-3,3,0.01)
    sy = [Likelihood(mu=mu, lam=lam).pdf(x) for x in sx]
    ax1.fill_between(sx, sy, label="likelihood", alpha=0.25, facecolor="orange")
    
    # Posterior of mean
    sx = np.arange(-3,3,0.01)
    sy = [PosteriorMean(data=data, m=m, lam=lam, beta=beta).pdf(x) for x in sx]
    ax1.fill_between(sx, sy, label="posterior of mean", alpha=0.25, facecolor="green")
    
    # Posterior of precision
    sx = np.arange(0,15,0.01)
    sy = [PosteriorPrecision(data=data, a=a, b=b, beta=beta, m=m).pdf(x) for x in sx]
    ax2.fill_between(sx, sy, label="posterior of precision", alpha=0.25, facecolor="green")

    # Predictive distribution
    m_hat = PosteriorMean(data=data, m=m, lam=lam, beta=beta).m_hat
    beta_hat = PosteriorMean(data=data, m=m, lam=lam, beta=beta).beta_hat
    a_hat = PosteriorPrecision(data=data, a=a, b=b, beta=beta, m=m).a_hat
    b_hat = PosteriorPrecision(data=data, a=b, b=b, beta=beta, m=m).b_hat

    sx = np.arange(-3,3,0.01)
    sy = [PredDist(m=m_hat, beta=beta_hat, a=a_hat, b=b_hat).pdf(x) for x in sx]
    ax1.fill_between(sx, sy, label="predictive distribution", alpha=0.25, color="tomato")
    
    # Surprise based on the predictive distribution
    sp_x = PredDist(m=m_hat, beta=beta_hat, a=a_hat, b=b_hat).surprise(data[-1])
    surprises.append(sp_x)
    sx = np.arange(len(data))
    sy = surprises
    ax3.plot(sx, sy, label="surprise", marker=".", color="tomato", alpha=0.8)
    
    # Error based on the predictive distribution
    err_x = PredDist(m=m_hat, beta=beta_hat, a=a_hat, b=b_hat).error(data[-1])
    errors.append(err_x)
    sx = np.arange(len(data))
    sy = errors
    ax4.plot(sx, sy, label="predictive error", marker=".", color="tomato", alpha=0.8)
    
    # Common settings
    ax1.legend(loc="upper left")
    ax1.set_xlabel("x")
    ax1.set_ylabel("Probability")
    ax1.set_title("Predictive distribution (N={})".format(len(data)), fontsize=12)
    
    ax2.legend()
    ax2.set_xlabel("x")
    ax2.set_ylabel("Probability")
    ax2.set_title("Variance (N={})".format(len(data)), fontsize=12)
    
    ax3.legend()
    ax3.set_xlabel("N: Number of observed data")
    ax3.set_ylabel("Surprise")
    ax3.set_title("Bayes surprise, Surprise={:.2f} (N={})".format(sp_x, len(data)), fontsize=12)
    
    ax4.legend()
    ax4.set_xlabel("N: Number of observed data")
    ax4.set_ylabel("Predictive error")
    ax4.set_title("Bayes predictive error, Error={:.2f} (N={})".format(err_x, len(data)), fontsize=12)
    

# Data sample
N = 500
mean = 1.2
precision = 10
data = [TrueDist(mean, precision).rvs(size=N)]
surprises = []
errors = []

# Figure settgins
fig = plt.figure(figsize = (12,12))
fig.suptitle("Bayes modeling of 1D data (Gauss-gamma distribution)", fontsize=16)
fig.subplots_adjust(hspace=0.3)

gs_master = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[2, 1, 1])
gs_1 = gridspec.GridSpecFromSubplotSpec(nrows=2, ncols=1, subplot_spec=gs_master[0, 0])
gs_2 = gridspec.GridSpecFromSubplotSpec(nrows=2, ncols=1, subplot_spec=gs_master[0, 1])
gs_3 = gridspec.GridSpecFromSubplotSpec(nrows=1, ncols=2, subplot_spec=gs_master[1, :])
gs_4 = gridspec.GridSpecFromSubplotSpec(nrows=1, ncols=2, subplot_spec=gs_master[2, :])
ax1  = fig.add_subplot(gs_1[:, :])
ax2  = fig.add_subplot(gs_2[:, :])
ax3  = fig.add_subplot(gs_3[:, :]) 
ax4  = fig.add_subplot(gs_4[:, :])

params = {
    'fig': fig, # Figureオブジェクト
    'func': update,  # グラフの描画関数
    'fargs': (data),  # 関数の引数 (フレーム番号を除く)
    'interval': 300,  # 更新間隔 (ms)
    'frames': np.arange(1, N+1, 1),  # フレーム番号を生成するイテレータ, 1~100
    'repeat': False,  # 繰り返さない
    'blit': False  # blitting による処理の高速化
}

# アニメーションを作成する 
ani = animation.FuncAnimation(**params)

# ani.save('bayes_surprise_error4.mp4', writer="ffmpeg")
plt.show()
```

```python
def prior_mean(x, m=0.0, lam=1, beta=1):    
    return stats.norm.pdf(x, loc=m, scale=1/(beta*lam))

def prior_precision(x, a=2, b=2):
    return scipy.stats.gamma.pdf(x, a, loc=0, scale=b)


def posterior_mean(x, data=None, m=None, lam=None, beta=None):
    N = len(data)
    m_hat = (np.sum(data) + beta * m) / (beta + N)
    beta_hat = beta + N
    return stats.norm.pdf(x, loc=m_hat, scale=1/(beta_hat*lam))


def posterior_precision(x, data=None, a=None, b=None):
    N = len(data)
    a_hat = a + N / 2
    b_hat = b + 0.5 * (np.sum(data ** 2) + beta * (m ** 2) - (1/(beta + N)) * (np.sum(data ** 2) + beta * m) ** 2)
    return stats.gamma.pdf(x, a_hat, loc=0, scale=b_hat)


def likelihood(x, mu=0.0, lam=1.0):
    return stats.norm.pdf(x, loc=mu, scale=1/lam)


def pred_dist(x, m=None, beta=None, a=None, b=None):
    df = 2 * a # degree of freedom
    loc = m
    scale = (beta/(beta + 1)) * (a/b)
    return stats.t.pdf(x, df, loc, scale)
```
